# Inspect content of checkpoint file

```python
import torch
```

```python
checkpoint_file = "C:/Users/z003y7sd/Downloads/last.ckpt"
```

```python
checkpoint = torch.load(checkpoint_file, map_location=torch.device('cpu'))

print("Epoch:", checkpoint.get("epoch", "Not available"))
print("Global Step:", checkpoint.get("global_step", "Not available"))
print("PyTorch-Lightning:", checkpoint.get("pytorch-lightning_version", "Not available"))
print("Checkpoint Keys:", checkpoint.keys())
```

```python
print("Model State Dict Keys:", checkpoint["state_dict"].keys())
```

```python
print("Model Loops Keys:", checkpoint["loops"].keys())
```

```python
print("Model Callbacks Keys:", checkpoint["callbacks"].keys())
```
